```python
import pandas as pd 
import numpy as np 
from sklearn.cluster import DBSCAN
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score

dataset = pd.read_csv('./csv_result-1222.csv')
dataset.head()
```

```python
# egg case 
from sklearn.preprocessing import StandardScaler 
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score 

X = dataset[dataset.columns].values
scaler = StandardScaler()
x_scaler = scaler.fit_transform(X)



#  AgglomerativeClustering instance
agg_clustering = AgglomerativeClustering(n_clusters=3)



# Fit the model to the standardized data
y_pred_agg = agg_clustering.fit(x_scaler)


calinski_harabasz_score_agg = calinski_harabasz_score(x_scaler,y_pred_agg.labels_)
print("calinski_harabasz_score dbscan:", calinski_harabasz_score_agg )

davies_bouldin_scores_agg = davies_bouldin_score(x_scaler,y_pred_agg.labels_)
print("davies_bouldin_score dbscan:", davies_bouldin_scores_agg) 


# print("Agglomerative Clustering Labels:")
# print(agg_clustering.labels_)
```

```python
# DBSCAN case 
from sklearn.preprocessing import StandardScaler
X = dataset[dataset.columns].values
scaler = StandardScaler()
x_scaler = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.5, min_samples=5)
y_pred_dbscan = dbscan.fit_predict(x_scaler)


from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score 

calinski_harabasz_score_dbscan = calinski_harabasz_score(x_scaler,y_pred_dbscan)
print("calinski_harabasz_score dbscan:", calinski_harabasz_score_dbscan )


davies_bouldin_scores = davies_bouldin_score(x_scaler,y_pred_dbscan)
print("davies_bouldin_score dbscan:", davies_bouldin_scores) 


 

```

```python
#kmean
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
y_pred_kmean = kmeans.fit(X) 
print(y_pred_kmean.labels_ )


from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score 

davies_bouldin_scores = calinski_harabasz_score(X,y_pred_kmean.labels_)
print("calinski_harabasz_score kmean:", davies_bouldin_scores) 



davies_bouldin_scores = davies_bouldin_score(X, y_pred_kmean.labels_)
print("davies_bouldin_scor:", davies_bouldin_scores) 


```
