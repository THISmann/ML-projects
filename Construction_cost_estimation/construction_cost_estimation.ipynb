Here we try to generate a fake dataset with a usefull information that we need for to estimate to cost in a project construction

```python
import pandas as pd
import random

# Liste des types de projet
project_types = ['Residential', 'Commercial', 'Industrial']

# Liste des états américains abrégés
states = ['kazan', 'Saint-peterburg', 'Moscow', 'Ekaterinburg', 'Tambov']

# Création d'un dataframe vide
df = pd.DataFrame(columns=[
    "Project_ID", "Description", "Type", "Location", "Site_Conditions",
    "Square_Footage", "Floors", "Units", "Foundation", "Framing",
    "Exterior_Finish", "Roofing", "HVAC", "Plumbing", "Electrical",
    "Fire_Protection", "Permitting_Fees"
])

# Génération de 10 lignes de données aléatoires
for i in range(1000):
    df.loc[len(df)] = {
        "Project_ID": i + 1,
        "Description": random.randint(1, 2),
        "Type": random.randint(1, 3),
        "Location": random.randint(1, 5),
        "Site_Conditions": random.randint(1, 10),
        "Square_Footage": round(random.uniform(1000, 100000)),
        "Floors": random.randint(1, 15),
        "Units": random.randint(1, 100),
        "Foundation": random.randint(1, 3),
        "Framing": random.randint(1, 3),
        "Exterior_Finish": random.randint(1, 3),
        "Roofing": random.randint(1, 3),
        "HVAC": random.randint(1, 2),
        "Plumbing": random.randint(1, 6),
        "Electrical": round(random.uniform(100, 1000)),
        "Fire_Protection": random.randint(0, 1),
        "Permitting_Fees": round(random.uniform(100, 5000), 2)
    }

# Afficher le contenu du dataframe
print(df)

# Enregistrement sous forme de fichier CSV
df.to_csv("construction_costs.csv", index=False)
```

Here we try to add a col for cost estimation with some parameters in the dataset

```python
# Ajout de la série 'cost_estimation'
dataCons['cost_estimation'] = (
    dataCons['Floors'] * 3000 + df['Units'] * 1000 +
    dataCons['Site_Conditions'].apply(lambda x: 8000 if x >  5 else 5000 if x < 0 else 0) +
    dataCons['Foundation'].apply(lambda x: 2000 if x == 'Slab' else 4000 if x == 'Crawlspace' else 2500 if x == 'Basement' else 0) +
    dataCons['Framing'].apply(lambda x: 4000 if x == 'Wood' else 3000 if x == 'Steel' else 3200 if x == 'Concrete' else 0) +
    dataCons['Exterior_Finish'].apply(lambda x: 3500 if x == 'Brick' else 2500 if x == 'Stucco' else 2500 if x == 'Siding' else 0) +
    dataCons['Roofing'].apply(lambda x: 4500 if x == 'Shingles' else 2500 if x == 'Tiles' else 3000 if x == 'Metal' else 0) +
    dataCons['HVAC'].apply(lambda x: 13500 if x ==
                           'gaz' else 8500 if x == 'electrique' else 0)
)

# Afficher le contenu du dataframe
print(dataCons)

# Enregistrement sous forme de fichier CSV
dataCons.to_csv("construction_costs.csv", index=False)
```

Here we separate the data in features and result

```python
X = dataCons[['Description', 'Type', 'Location', 'Site_Conditions', 'Square_Footage', 'Floors', 'Units', 'Foundation', 'Framing', 'Exterior_Finish', 'Roofing', 'HVAC', 'Plumbing', 'Electrical', 'Fire_Protection', 'Permitting_Fees']]
y = dataCons['cost_estimation']
```

```python
X
```

Here we apply a lenear regression model on the dataset

```python
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

here we put same data to test the model

```python
new_input = np.array([[1, 3, 1,7,50000,10,34,2,2,1,2,1,1,3,1,10000]])  # New input values
cost_estimation = regressor.predict(new_input)
print("Predicted Cost:", cost_estimation)
```

```python
k1 = regressor.intercept_
print("Intercept (k1):", k1)
```

```python
r2_score = regressor.score(X_test, y_test)
print("R² Score:", r2_score)
```
